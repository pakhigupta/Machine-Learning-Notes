```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
#Simple Example
x = np.arange(1, 10)
y = np.arange(11, 20)
print("x", x)
print("y", y)
```

```python
a = np.arange(40, 50)
b = np.arange(50, 60)
print("a", a)
print("b", b)
```

## Plotting using matplotlib

### Scattering 

```python
# plt scatter
plt.scatter(x, y, c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('ScatteredGraph.png')
```

```python
y = x*x
```

### Plotting 

```python
# plt plotting
plt.plot(x, y, 'b*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

### Subplots 

```python
# creating subplots
plt.subplot(2,2,1)
plt.plot(x,x,'r')
plt.subplot(2,2,2)
plt.plot(x,1/x,'b')
plt.subplot(2,2,3)
plt.plot(x,x*x*x,'g')
```

```python
# sin and cosine curves

x = np.arange(-2, 18, 0.01)

y1 = np.sin(x)
plt.subplot(2, 1, 1)
plt.plot(x, y1, 'r--')
plt.title('Sin Curve')

y2 = np.cos(x)
plt.subplot(2, 1, 2)
plt.plot(x, y2, 'b--')
plt.title('Cos Curve')
```

## Plotting a bar

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11,5]
y2 = [6,15,7,10]

plt.bar(x, y, align = 'center')
plt.bar(x2, y2, color = 'g')

plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

### Histograms

```python
a = np.random.randn(50)*100
plt.hist(a, bins=20)
plt.title('Histogram')
```

## Box Plot using Matplotlib  

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

#rectangular box plot
plt.boxplot(data, vert=True, patch_artist=True)
```

## Pie Charts 

```python
labels = ['Python', 'C', 'C++', 'Java']
sizes = [215, 100, 250, 150]
colors = ['gold', 'lightcoral', 'yellowgreen', 'lightskyblue']
explode = (0.1, 0, 0, 0) #how far this slice will go from the chart

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)

#autopct = format of the decimat values in decimal
```
