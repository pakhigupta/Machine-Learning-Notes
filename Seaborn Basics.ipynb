```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.dtypes
```

# Correlation

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# JoinPlot
### 2 features

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

# PairPlot
### more than 2 features

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='smoker')
```

# Dist Plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

# Categorical Plot
#### Boxplot, violinplot, countplot,  barplot

# Count Plot

```python
sns.countplot('day', data=df)
```

```python
sns.countplot(y='day', data=df)
```

# Bar plot

```python
sns.barplot(x='tip', y='day', data=df)
```

# Box Plot

```python
sns.boxplot('smoker', 'total_bill', data=df)
```

```python
sns.boxplot(data=df, palette='rainbow')
```

```python
sns.boxplot('smoker', 'total_bill', data=df, orient='v', hue='day')
```

# Violin Plot

```python
sns.violinplot(x='sex', y='total_bill', data=df, hue='day')
```
